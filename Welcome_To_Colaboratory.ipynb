```
#preprocessing



# Here we get the length of each review
data['Length']= data['Review'].apply(len) 
data.head()
```

```
import re
import nltk
nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
corpus = []
for i in range(0, 1000):
   # Cleaning special character from the reviews
    review = re.sub('[^a-zA-Z]', ' ', data['Review'][i])
    # Converting the entire review into lower case
    review = review.lower()
     # Tokenizing the review by words
    review = review.split()
       # Stemming the words
    ps = PorterStemmer()
      
    # Removing the stop words
    review = [ps.stem(word) for word in review if not word in set(stopwords.words('english'))]
    # Joining the stemmed words
    review = ' '.join(review)
    corpus.append(review)
   
```

```
#From the cleaned dataset, potential features are extracted and are converted to,
 #numerical format. The vectorization techniques are used to convert textual 
 #data to numerical format. Using vectorization, a matrix is created where
  #each column represents a feature and each row represents an individual review.
  
    # bag of words

from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 1500)
X = cv.fit_transform(corpus).toarray()
y = data.iloc[:, 1].values 
```
